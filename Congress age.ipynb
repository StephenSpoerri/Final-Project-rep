```python
import matplotlib.pyplot as plt
import numpy as np
Congress = [93,98,103,108,113]
Age = [50,54,58.2,59.3,61.3]
df = pd.DataFrame(list(zip(Congress, Age)),
               columns =['Congress', 'Age'])
print(df)
df.plot(kind='scatter', y='Age', x='Congress')
df.plot(kind='line', y='Age', x='Congress')
plt.xlabel('Congress')
plt.ylabel('Age')
plt.show()
X = df['Congress']
y = df['Age']


theta = np.polyfit(X, y, 1)

print(f'The parameters of the curve: {theta}')


y_line=theta[1] + theta[0] * X

plt.scatter(X, y)
plt.plot(X, y_line, 'r')
#graphs for average age of senators over time
```

```python
#https://www.statista.com/statistics/984745/youth-voter-turnout-presidential-elections-us/
Year = [1968, 1972, 1976, 1980, 1984, 1988, 1992, 1996, 2000, 2004, 2008, 2012,2016]
Turnout = [60.7, 55.4, 48.8, 48.2, 49.1, 43.8, 52, 39.6, 40.3, 49, 51.1, 45, 44]
df = pd.DataFrame(list(zip(Year, Turnout)),
               columns =['Year', 'Turnout'])
print(df)
df.plot(kind='scatter', y='Turnout', x='Year')
df.plot(kind='line', y='Turnout', x='Year')
plt.xlabel('Year')
plt.ylabel('Turnout')
plt.show()
X = df['Year']
y = df['Turnout']


theta = np.polyfit(X, y, 1)

print(f'The parameters of the curve: {theta}')


y_line=theta[1] + theta[0] * X

plt.scatter(X, y)
plt.plot(X, y_line, 'r')
#graphs for average voter turnout for youths 
```
